```python
estadisticas_descriptivas = df.describe()

moda = df.mode().iloc[0]

print("Estadísticas Descriptivas:")
print(estadisticas_descriptivas)
print(f"Moda: \n{moda}")
```

```python
plt.figure(figsize=(10, 6), dpi=300)
plt.hist(df['Valores'], bins=range(min(datos), max(datos) + 2, 1), edgecolor='black', alpha=0.7)
plt.title('Histograma de los Valores')
plt.xlabel('Valores')
plt.ylabel('Frecuencia')
plt.xticks(range(0, 15 + 1, 1))
plt.yticks(range(0, 40 + 1, 5))
plt.grid(axis='y', alpha=0.75)
plt.grid(axis='x', alpha=0.75)
```
