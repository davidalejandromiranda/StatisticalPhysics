## 1. Librerías y métodos 

```python
import numpy as np 
import matplotlib.pyplot as plt 
import pandas as pd
```

```python
def plot_with_error(df, x_label, param_label, linreg=True, instrument_error=0, marker='', color='', label=''):
    x = []
    y_mean = []
    y_std  = []
    for xk, df in df.groupby(x_label):
        x.append(xk)
        y_mean.append(np.mean(df[param_label]))
        y_std.append(max([instrument_error, np.std(df[param_label])]))
    for k in range(len(x)):
        y1 = y_mean[k] - y_std[k]
        y2 = y_mean[k] + y_std[k]
        plt.plot(2*[x[k]], [y1, y2], '+-'+color, alpha=0.8, lw=1, ms=4)
    if not linreg:
        plt.plot(x, y_mean, marker+color, label=label)
    if linreg:
        p = np.polyfit(x, y_mean, 1)
        label_linreg = 'Pendiente: %0.0f, Intercepto: %0.1f' % (p[0], p[1])
        plt.plot(x, np.polyval(p, x), color, label=label_linreg)
        plt.legend()

    plt.xlabel(x_label)
    plt.ylabel(param_label)
```
