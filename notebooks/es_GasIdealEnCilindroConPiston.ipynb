## 2. Planteamiento del problema

Un contenedor cilíndrico con área transversal $A$ y longitud $L$ está separado en dos compartimentos por un pistón delgado, inicialmente fijo a una longitud $l_a$, medida desde el extremo izquierdo del cilindro. El compartimento de la izquierda contiene $\nu_a$ moles de helio gaseoso a una presión de $p_a$ atmósferas; el compartimento de la derecha está lleno con gas argón a $p_b$ atmósferas de presión. Estos gases se pueden considerar ideales.  El cilindro es sumergido en $L_{H_2O}$ litros de agua y el sistema entero está inicialmente a temperatura ambiente, 25ºC. Las capacidades caloríficas del cilindro y el pistón pueden ser despreciadas.  Cuando se retira el pistón, después de un cierto tiempo, se alcanza el estado de equilibrio cuando el pistón alcanza una nueva posición.

a) ¿Cuál es la temperatura del agua cuando el sistema alcanza nuevamente el equilibrio?

c) ¿Cuánto es la variación de la entropía de todo el sistema para diferentes posiciones iniciales?

d) ¿Cómo depende la fracción molar $\nu_b$ de la posición inicial del pistón?

## 4. Resultados
### 4.1. Condiciones iniciales

```python
L = 80 # cm
l_a = np.linspace(0.1*L, 0.9*L, 100) 
nu_a = 1 # mole
p_a = 5 # atm
p_b = 1 # atm
```

### 4.2. Cálculo de variables

```python
l_b = L - l_a
xi = p_a*l_a/(p_b*(L-l_a))
nu_b = nu_a / xi
laf = xi*L/(1+xi)
lbf = L - laf
DS = nu_a * np.log(laf/l_a) + nu_b * np.log(lbf/l_b)
```

### 4.3. Posición final en función de la posición inicial del pistón

```python
plt.figure(dpi=300)
plt.plot(l_a, laf)
plt.xlabel('$l_a^{(inicial)}$ [cm]')
_ = plt.ylabel('$l_a^{(final)}$ [cm]')
```

### 4.4. Variación de la entropía total en función de la posición inicial del pistón

```python
plt.figure(dpi=300)
plt.plot(l_a, DS)
plt.xlabel('$l_a^{(inicial)}$ [cm]')
_ = plt.ylabel(r'$\frac{\Delta S_T}{R}$  [mole]')
```
