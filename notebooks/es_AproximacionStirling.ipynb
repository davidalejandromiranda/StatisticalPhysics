La [aproximación de Stirling](https://en.wikipedia.org/wiki/Stirling%27s_approximation) permite estimar el factorial $N!$ de números grandes.  El logaritmo natural de $N!$ está dado por:

$$ln(N!) = N ln(N) - N + \mathcal{O}\{ ln(N) \}$$

Al expresar en base 2 el logaritmo y tomando el término $\frac{1}{2}ln(2\pi n)$ como la contribución del error de orden superior, se obtiene:

$$N! \approx \sqrt{2\pi N} \left( \frac{N}{e} \right)^N $$

Es importante notar que la función $\Gamma(z)$ generaliza el concepto de factorial para cualquier número complejo $z$, tal que para el número natural $N$, su factorial será,

$$N! = \Gamma(N+1)$$

$$\Gamma(z) = \int_0^\infty t^{z-1}e^{-t} dt $$

## 1. Cargar librerías

```python
import numpy as np 
from scipy.special import gamma
import matplotlib.pyplot as plt 
```

## 2. Aproximación de $N!$

```python
N = np.arange(1, 5) 
n = np.linspace(min(N), max(N), 1000)
N_fact = [np.math.factorial(n) for n in N]
Stirling0 = np.exp(n * np.log(n) - n)
Stirling1 = np.sqrt(2*np.pi*n)*(n/np.exp(1))**n
plt.figure(dpi=300)
plt.plot(N, N_fact, 'ok', label='N!')
plt.plot(n, gamma(n+1), ':g', label=r'$\Gamma(n+1)$')
plt.plot(n, Stirling1, ':r', label=r'$(2\pi n)^{1/2} \left( \frac{n}{e} \right)^n $')
plt.plot(n, Stirling0, '--b', label=r'$e^{ n ln(n) - n }$')
plt.xlabel('n')
plt.legend()
_ = plt.title('Figura 1')
```

## 3. Aproximación de $ln(N!)$

```python
N = np.arange(1, 100, 20) 
n = np.linspace(min(N), max(N), 1000)
N_fact = [np.math.factorial(n) for n in N]
log_N  = [np.log(float(n_fact)) for n_fact in N_fact]
ln_Stirling0 = n * np.log(n) - n
ln_Stirling1 = np.log(np.sqrt(2*np.pi*n)*(n/np.exp(1))**n)
plt.figure(dpi=300)
plt.plot(N, log_N, 'ok', label=r'$ln(N!)$')
plt.plot(n, np.log(gamma(n+1)), 'g', label=r'$ln\{\Gamma(n+1)\}$')
plt.plot(n, ln_Stirling1, ':r', label=r'$ln\left\{(2\pi n)^{1/2} \left( \frac{n}{e} \right)^n \right\} $')
plt.plot(n, ln_Stirling0, '--b', label=r'$n ln(n) - n$')
plt.xlabel('n')
plt.legend()
_ = plt.title('Figura 2')
```
