## 2. Planteamiento del problema

En 2016, [Miranda y Bueno](https://pubs.rsc.org/en/content/articlelanding/2016/CP/C6CP01659F) reportaron un funcional de la densidad electrónica para sistemas mesoscópicos electroactivos. En [2019](https://pubs.acs.org/doi/10.1021/acs.jpcc.9b04020), mostraron que el funcional reportado permitía describir datos experimentales a temperaturas finitas, con lo cual su uso se extendió para analizar reactividad química. En 2021, [Corzo, Bueno y Miranda](https://doi.org/10.1109/ACCESS.2021.3135787) reportaron la dureza química efectiva, definida en términos del funcional Miranda-Bueno, como elemento conceptual relevante para mejorar la reproducibilidad de datos experimentales de sensores cuyo principio de funcionamiento es la capacitancia cuántica.  En este *Worked Example* se muestra un análisis de los resultados reportados en 2021, [Corzo et. al. 9 (2021) 166446-166454](https://doi.org/10.1109/ACCESS.2021.3135787), para ello se analizan los datos contenidos en la [información complementaria](https://ieeexplore.ieee.org/ielx7/6287639/9312710/9650898/supp1-3135787.docx?arnumber=9650898) del artículo.

### 2.1. El funcional Miranda-Bueno

La teoría del funcional de la densidad electrónica, [DFT](https://en.wikipedia.org/wiki/Density_functional_theory), ha tenido un gran desarrollo desde que Khon y Shan desarrollaron métodos númericos que permitieron su implementación para estudios de materiales. Esta teoría se basa en los teorema de Hohenberg-Kohn, en los cuales se muestra que la energía de un sistema es un funcional de la densidad electrónica y la energía obtenida a partir de dicho funcional corresponde con la misma predicha por la ecuación de Schrödinger. En tal sentido, el funcional Miranda-Bueno fue propuesto para estudiar sistemas mesoscópicos electroactivos; dicho funcional se basa en las características capacitivas de dichos sistemas, lo cual permite definir la energía en términos de la densidad electrónica y la capacitancia electroquímica. Es importante tener presente que la capacitancia electroquímica depende de la densidad electrónica, de hecho, es un funcional de la densidad electrónica y se puede obtener experimentalmente.

Para un sistema electroactivo con hamiltoniano $\hat{H}$, en la base de estados naturales del sistema $|\psi_n\rangle$, la energía $E = E_\mu[\rho]$ está dada en términos de la carga del electrón $q_e$, el volumen ocupado por el sistema $\mho$, la densidad electrónica $\rho$ y la capacitancia electroquímica $C_\mu[\rho] = zq_e \frac{\partial q}{\partial \mu} $, donde $z$ es el número de electrones que se pueden intercambiar y $\mu$, el potencial químico.

$$E_\mu[\rho] = \langle \psi_n | \hat{H} | \psi_n \rangle = \frac{1}{2}\frac{[ z q_e  \rho(\mu)\mho]^2 }{C_\mu[\rho]}$$

### 2.2. El funcional a temperatura absoluta $T > 0$

En 2019, Miranda-Bueno mostraron que el funcional $E_\mu[\rho]$ corresponde con el gran potencial $\Omega = -\frac{1}{\beta} ln \left\{ tr \left[ e^{-\beta(\hat{H} - \mu \hat{N})} \right] \right\}$ aprovechando la definición dada en el ítem anterior. De acuerdo a la definición del funcional Miranda-Bueno, solo los elementos de la traza del hamiltoniano contribuyen con la energía, por lo tanto, al evaluar $ln \left\{ tr \left[ e^{-\beta(\hat{H} - \mu \hat{N})} \right] \right\}$ se obtiene $-\beta E_\mu[\rho]$, con lo cual se demuestra que $E_\mu[\rho] = \Omega$. La implicación directa de este resultado es que el funcional Miranda-Bueno permite describir, con la misma estructura matemática, la energía de sistemas electroactivos para cualquier temperatura. Esto permite que con resultados experimentales a temperatura ambiente se pueda obtener la expresión de la energía del sistema a partir de medidas macroscópicas de capacitancia electroquímica.

### 2.3. La dureza química

Rober Parr y Ralph Pearson definieron, En términos de la energía $E[\rho]$ del sistema, la dureza química $\eta$ como,

$$\eta = \frac{\delta^2 E[\rho]}{\delta \rho^2}$$

En términos del funcional Miranda-Bueno, la dureza química se expresa como:

$$\eta = \frac{2\gamma'}{C_{\bar{\mu}}} - \frac{8\gamma'^2\rho}{C^3_{\bar{\mu}}} \frac{\delta{C_{\bar{\mu}}}}{\delta{\bar{\mu}}} + \frac{12\gamma'^3\rho^2}{C^5_{\bar{\mu}}} \left( \frac{\delta{C_{\bar{\mu}}}}{\delta{\bar{\mu}}} \right)^2 - \frac{4\gamma'^3\rho^2}{C^4_{\bar{\mu}}} \frac{\delta^2{C_{\bar{\mu}}}}{\delta{\bar{\mu}^2}}$$

Para sistemas electroactivos con valores de capacitancia relativamente altos ($C > 1 \mu F$), domina el primer término, por lo tanto, tomando $\gamma = 2\gamma'= (z q_e  \mho)^2$,

$$\eta \approx \frac{\gamma}{C_{\bar{\mu}}}$$

### 2.4. Dureza química efectiva

Sea un sistema electroactivo sin contacto con analito (blanco), cuya densidad electrónica está dada por $\rho_b$ y la capacitancia cuántica por $C_{q,0}$. Al colocar en contacto al sistema electroactivo con analito, la densidad electrónica es $\rho$ y la capacitancia, $C_q$.  Se puede asumir que el efecto neto del analito está dado por $\Delta \rho = \rho_b - \rho$, con lo cual, la capacitancia efectiva es $\Delta C = C_{q,0} - C_q$. Como la dureza química es $\eta = \frac{\delta^2 E[\rho]}{\delta \rho^2}  \approx \frac{\gamma}{C_{\bar{\mu}}}$, entonces, se puede definir la dureza química efectiva como,

$$\eta_e = \frac{\delta^2 E[\Delta \rho]}{\delta \rho^2}  \approx \frac{\gamma}{\Delta C} = \frac{\gamma}{C_{q,0} - C_q}$$

Siguiendo las recomendaciones de la IUPAC de restar la señal del blanco, se define,

$$\frac{\eta_e^*}{\gamma} = \frac{1}{C_{q,0} - C_q} - \frac{1}{C_q}$$

## 3. Métodos

```python
def get_samples_labels(sample_name):
    samples = []
    for this_label in data.Sample.unique():
        if sample_name in this_label:
            samples.append(this_label)
    return samples
```

```python
def get_data_by(sample_name):
    samples = get_samples_labels(sample_name)
    mask_num = sum([data.Sample == s for s in samples])
    mask = [m == 1 for m in mask_num]
    these_samples = data.loc[mask]
    return these_samples
```

```python
def capacitance(these_samples):
    desc = these_samples.describe()
    x = list(desc.columns)
    C_mean = desc.loc['mean'].to_numpy()
    C_std = desc.loc['std'].to_numpy()
    return x, C_mean, C_std
```

```python
def hardness(these_samples):
    x = list(these_samples.describe().columns)
    η = these_samples[x].apply(lambda row: 1/row - 1/row[0], axis=1)
    desc = η.describe()
    η_mean = desc.loc['mean'].to_numpy()
    η_std = desc.loc['std'].to_numpy()
    return x, η_mean, η_std
```

```python
def effective_hardness(these_samples):
    x = list(these_samples.describe().columns)
    ηe = these_samples[x].apply(lambda row: 1/(row[0]-row) - 1/row[0], axis=1)
    x.remove(0)
    desc = ηe[x].describe()
    ηe_mean = desc.loc['mean'].to_numpy()
    ηe_std = desc.loc['std'].to_numpy()
    return x, ηe_mean, ηe_std
```

```python
def get_xye_dict(fun):
    xye_dict = {}
    for sample_name in ['a', 'b', 'c']:
        these_samples = get_data_by(sample_name)
        x, mean, std = fun(these_samples)
        xye_dict[sample_name] = {
            'x': x,
            'mean': mean,
            'std': std,
        }
    return xye_dict
```

```python
def plot(fun, ylabel=''):
    xye_dict = get_xye_dict(fun)
    plt.figure(dpi=300)
    colors = 'kbr'
    for k, sample_name in enumerate(xye_dict.keys()):
        xx = xye_dict[sample_name]['x']
        y = xye_dict[sample_name]['mean']
        yerr = xye_dict[sample_name]['std']
        plt.errorbar(
            xx,
            y,
            yerr=yerr, 
            fmt='.',
            label=sample_name,
            ecolor=colors[k], color=colors[k])
        plt.gca().set_xscale('log')
    plt.xlabel('[x]  /  $\mu M$')
    plt.ylabel(ylabel)
    plt.legend()
```
