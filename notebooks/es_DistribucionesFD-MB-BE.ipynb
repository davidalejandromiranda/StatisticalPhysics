## 1. Importar librerías

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.constants as ct
```

## 2. Definiciones y métodos

```python
T = 300 # K
β = 1 / ( ct.Boltzmann * T )
E = np.linspace(0, 10/β, 2000)
```

```python
def dist(β, μ):
    A = 1 / ( np.exp( β * ( E - μ ) ) + 1 )
    A /= max(A)
    B = np.exp(-β * E)
    B /= max(B)
    C = 1 / ( np.exp( β * ( E - μ ) ) - 1 )
    C /= max(C)
    return [A, B, C]
```

```python
def plot_dist(β, μ):
    for label, y in zip(['A', 'B', 'C'], dist(β, μ)):
        plt.semilogx(E/ct.elementary_charge, y, label=label, alpha=0.7)
        plt.yticks([])
    plt.vlines(x = μ/ct.elementary_charge, ymin=0, ymax=1, linestyles=':', color='k', lw=2)
    plt.ylim(0, 1)
    plt.xlabel('Energía [eV]')
    plt.legend()
```

## 3. Figuras

```python
plt.figure(dpi=300)
plot_dist(β, 1/β)
_ = plt.title('Figura 1')
```

```python
plt.figure(dpi=300)
labels = ['a', 'b', 'c', 'd']
chem_potentials = np.array([0.1, 0.5, 1, 5])/β
for i, (label, μ) in enumerate(zip(labels, chem_potentials)):
    plt.subplot(2,2,i+1)
    plot_dist(β, μ)
    _ = plt.title('Figura 2%s' % label)
plt.tight_layout()
```
